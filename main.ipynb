```python
from turtle import *
import time
import random
```

```python
score=0
execution_delay=0.1

root=Screen()#object of screen class
root.title('Snake Game')
root.bgcolor('black')
root.setup(width=600,height=600)
root.bgpic('border.gif')
root.tracer(False)
root.addshape('upmouth.gif')
root.addshape('food.gif')
root.addshape('downmouth.gif')
root.addshape('leftmouth.gif')
root.addshape('rightmouth.gif')
root.addshape('body.gif')


head=Turtle()
head.shape('upmouth.gif')
head.penup() #no line
head.goto(0,0)  #place of snake
head.direction='stop'




food=Turtle()
food.shape('food.gif')
food.penup()
food.goto(0,100)
food.pensize(1)
execution_delay=execution_delay-0.003





text=Turtle()
text.penup()
text.goto(0,268)
text.hideturtle()
text.color('white')
text.write('Score:0',font=('courier',25,'bold'),align='center')


lost=Turtle()
lost.color('white')
lost.penup()
lost.hideturtle()
segments=[]   
# update_frame



def exist_in_seg(x,y):
    if len(segments) == 0: return False
    for i in segments:
        if x == i.xcor() and y == i.ycor(): return True
    return False

def valid(dir,next_dir,x , y):
    if opp_dir(dir,next_dir): return True
    if x < -260 or x > 260: return False
    if y < -260 or y > 260: return False
    if exist_in_seg(x,y) :return False
    return True

def getNext(dir,curx,cury):
    if dir == "up": return(curx,cury+20)
    if dir == "down": return (curx,cury-20)
    if dir == "right": return (curx+20,cury)
    if dir == "left": return (curx-20,cury)



def opp_dir(dir,cur_dir):
    if (dir == "down" and cur_dir == "up") or (dir == "up" and cur_dir == "down"):return True
    if (dir == "up" and cur_dir == "down") or (cur_dir == "up" and dir == "down"):return True
    if (dir == "left" and cur_dir == "right") or (dir == "right" and cur_dir == "left"):return True
    if (dir == "right" and cur_dir == "left") or (dir == "left" and cur_dir == "right"):return True
    return False




def update_frame():

    dir = ["up","left","right","down"]
    if head.direction == "stop": return
    head.shape(str(head.direction)+"mouth.gif")
    if food.xcor() == head.xcor():
        if food.ycor() > head.ycor():head.direction = "up"
        else: head.direction = "down"

    if food.ycor() == head.ycor():
        if food.ycor() > head.ycor():head.direction = "right"
        else: head.direction = "left"

    x , y = getNext(head.direction,head.xcor(),head.ycor())
    if not valid(head.direction,head.direction,x,y):
        for k in dir:
            if opp_dir(k,head.direction): continue
            i,j = getNext(k,head.xcor(),head.ycor())
            if valid(head.direction,k,i,j): 
                head.direction = k
                break






    if head.direction=='up':
        y=head.ycor()
        y=y+20
        head.sety(y)
        
    if head.direction=='down':
        y=head.ycor()
        y=y-20
        head.sety(y)
        
    if head.direction=='right':
        x=head.xcor()
        x=x+20
        head.setx(x)
        
    if head.direction=='left':
        x=head.xcor()
        x=x-20
        head.setx(x)  
        
        
        
        
##################################################################        
        
def go_up():
    if head.direction!='down':
        head.direction='up'
        head.shape('upmouth.gif')

def go_down():
    if head.direction!='up':
        head.direction='down'
        head.shape('downmouth.gif')


def go_left():
    if head.direction!='right':
        head.direction='left'
        head.shape('leftmouth.gif')


def go_right():
    if head.direction!='left':
        head.direction='right'
        head.shape('rightmouth.gif')


        
        
root.listen()

root.onkeypress(go_up,'Up')
root.onkeypress(go_down,'Down')
root.onkeypress(go_left,'Left')
root.onkeypress(go_right,'Right')
     
        
######################################################################################################################





while True:
    root:update()
    
        
        
    if head.xcor()>260 or head.xcor()<-260 or head.ycor()>260 or head.ycor()<-260:
        lost.write('Game Lost',align='center',font=('courier',34,'bold'))
        time.sleep(1)
        lost.clear()
        time.sleep(1)
        head.goto(0,0)
        head.direction='stop'
        for bodies in segments:
            bodies.goto(1000,1000)
        
        
    if head.distance(food)<20:
        
        x=random.randint(-255,255)
        y=random.randint(-255,255)

        while x % 20 != 0: x-= 1
        while y % 20 != 0: y-= 1

        while exist_in_seg(x,y):
            x = random.randint(-255,255)
            y = random.randint(-255,255)
            while x % 20 != 0: x-= 1
            while y % 20 != 0: y-= 1


            

        food.goto(x,y)
            

        ##execution_delay=execution_delay-0.003
        
        
        body=Turtle()
        body.penup()
        body.shape('body.gif')
        segments.append(body)
        score=score+10
        text.clear()
        
        text.write(f'Score:{score}',font=('courier',25,'bold'),align='center')
        
        
    for i in range(len(segments)-1,0,-1):
        x=segments[i-1].xcor()
        y=segments[i-1].ycor()
        segments[i].goto(x,y)
        
        
    if len(segments)>0:
        x=head.xcor()
        y=head.ycor()
        segments[0].goto(x,y)
    
        
    update_frame()
    
    for bodies in segments:
        if bodies.distance(head)<20:
            time.sleep(1)
            head.goto(0,0)
            head.direction='stop'

            for bodies in segments:
                bodies.goto(1000,1000)

            segments.clear()
            score=0
            execution_delay=0.1
            lost.write('Game Lost', align='center', font=('courier', 34, 'bold'))
            time.sleep(1)
            lost.clear()

            text.clear()
            text.write('Score:0', align='center', font=('courier', 25, 'bold'))

    time.sleep(execution_delay)

    
        
```
